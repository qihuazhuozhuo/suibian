```python
starttime = list(df['starttime'])
stoptime = list(df['stoptime'])
```

```python
#strftime是把日期格式转化为字符串格式
#strptime是把字符串格式转化为日期格式
def change_datetime(t):
    try:
        t1 = datetime.strptime(t, '%m/%d/%Y %H:%M')
    except:
        t1 = datetime.strptime(t, '%m/%d/%Y %H:%M:%S')
    return t1
```

```python
df['start_time'] = df['starttime'].apply(change_datetime)#将此函数应用到此列
df['stop_time'] = df['stoptime'].apply(change_datetime)
```

```python
ind = pd.DatetimeIndex(df.start_time) #不可变的datetime64数据的n维数组，内部表示为int64
df['date'] = ind.date.astype('datetime64')
df['hour'] = ind.hour
```

```python
by_date = df.pivot_table('tripduration', aggfunc='count', 
                         index = 'date',
                        columns = 'usertype',)
```

```python
by_date.head()
```

```python
fig,ax = plt.subplots(2, figsize=(16,8))
fig.subplots_adjust(hspace = 0.4)#子图之间的相距高度
by_date.iloc[:,0].plot(ax=ax[0], title='Customer');#iloc：纯粹的基于整数位置的索引按位置选择。
by_date.iloc[:,1].plot(ax=ax[1], title='Subscriber')
```

<font color=red size=5> 很明显，每周都有自行车使用模式（数据中每月有4-5个峰值）。 8月份，customer数量高峰，可能是因为夏季天气好。 游客何时主要访问纽约市？ 从5月到9月，subscriber的数量已经达到了稳定的模式; 在劳动节周末，夏季结束时的数据中有一些有趣的结构。 用户发起的游乐设施数量在劳动节周末之后是最高的，也许是因为更多人在此日期之后订购了该服务？ </font>

```python
list(set(by_date.index.dayofweek)) #Mondays = 0；set() 函数创建一个无序不重复元素集
```

```python
by_weekday = by_date.groupby(by_date.index.dayofweek).mean()#以星期为单位，用户分类统计平均每天骑行时间
by_weekday.columns.name = None
by_weekday.plot(title = 'Average use by day of week')
```

```python
by_weekday
```

```python
by_weekday['ratio'] = by_weekday['Subscriber']/by_weekday['Customer']
by_weekday
```

- Subscriber平均使用自行车的次数比Customer多得多。
- 尽管Subscriber在工作日期间使用自行车的次数最多，但Customer在周末会更多地使用自行车

```python
#按整时统计
by_hour = df.pivot_table('tripduration', aggfunc = 'count',
                        index=['date','hour'], 
                        columns = 'usertype').reset_index('hour')
#reset_index('hour')使多级索引变为以date为索引
```

```python
by_hour.tail()
```

```python
by_hour.index
```

```python
by_hour['weekend'] = (by_hour.index.dayofweek >=5)
```

```python
by_hour.head()
```

```python
#按工作日、周末分类，按整时统计，并求均值
by_hour = by_hour.groupby(['weekend','hour']).mean()
```

```python
by_hour
```

```python
#format():格式化字符串的方法，0代表第一个参数
by_hour.index.set_levels([['weekday', 'weekend'],
                         ["{0}:00".format(i) for i in range(24)]],
                        inplace = True)
by_hour.columns.name = None
```

```python
by_hour
```

```python
fig, ax = plt.subplots(1, 2, figsize=(16,6), sharey=True)#sharey=True:x或y轴将在所有子图中共享。
by_hour.loc['weekday'].plot(title='Avg hourly use, weekdays', ax = ax[0])
by_hour.loc['weekend'].plot(title='Avg hourly use, weekends', ax =ax[1])
ax[0].set_ylabel('Avg trips per hour')
#loc:纯粹基于标签位置的索引器，用于按标签进行选择。
```

- Subscriber在工作日和周末使用自行车的时间不同
- 平日通勤时间内平均旅行次数更多！
- Subscriber工作日的使用受到通勤的驱动：Subscriber使用高峰期在工作日的清晨和傍晚。
- 另外三条都是在下午的早些时候达到了高峰

```python
# Same analysis; different aggregating function: count instead of average,看图两个情况相同
by_hour = df.pivot_table('tripduration', aggfunc = 'count',
                        index=['date','hour'], 
                        columns = 'usertype').reset_index('hour')

by_hour['weekend'] = (by_hour.index.dayofweek >=5)
by_hour = by_hour.groupby(['weekend','hour']).sum()
by_hour
```
